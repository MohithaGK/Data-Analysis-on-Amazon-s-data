# Group by 
this is used to group the data based on one or more cloumns in a db.

```python
"""This line of code performs several operations on a DataFrame `df`:

1. `df.groupby(['Size'], as_index=False)`: This groups the DataFrame `df` by the column 'Size'. The `as_index=False`
parameter ensures that the grouped column ('Size' in this case) is not set as the index of the resulting DataFrame. 
Instead, it remains as a regular column.

2. `['Qty'].sum()`: After grouping by 'Size', it selects the column 'Qty' and calculates the sum of 
its values within each group. This operation aggregates the data for each unique 'Size' value, summing up the 'Qty' 
values for each group.

3. `.sort_values(by='Qty', ascending=False)`: Once the grouping and summation are done, this sorts the resulting 
DataFrame based on the sum of 'Qty' values in descending order. The `by='Qty'` parameter specifies that we want
to sort by the 'Qty' column, and `ascending=False` ensures that the sorting is done in descending order.

In summary, this line of code groups the DataFrame by the 'Size' column, calculates 
the sum of 'Qty' values within each group, and then sorts the groups based on the sum of 'Qty' values 
in descending order. The `as_index=False` parameter ensures that the grouped column ('Size') remains as a 
regular column in the resulting DataFrame.
So,here we get ouput as in series..not in df..so to get in df we put reset_index() at last to get in df ouput..the syntax is
like this
df.groupby(['Size'])['Qty'].sum().sort_values(ascending=False).reset_index()"""

df.groupby(['Size'])['Qty'].sum().sort_values(ascending=False)

#df.groupby(['Size'], as_index=False)['Qty'].sum().sort_values(by='Qty',ascending=False)
```

```python
"""The error occurs because whenyou pass the grouped and sorted data (bars) directly to the data
parameter of sns.barplot(), Seaborn cannot interpret the input as it expects a DataFrame, not a Series.

To fix this, you should reset the index of the bars Series to convert it back into a
DataFrame before passing it to sns.barplot()"""

bas = df.groupby(['Size'])['Qty'].sum().sort_values(ascending = False).reset_index()
sn.barplot(x = 'Size', y = 'Qty',data = bas)
```

```python
top_10_state = df['ship-state'].value_counts().head(10)
# Plot count of cities by state
plt.figure(figsize=(12, 8))
sn.countplot(data=df[df['ship-state'].isin(top_10_state.index)], x='ship-state')
plt.xlabel('ship-state')
plt.ylabel('count')
plt.title('Distribution of  State')
plt.xticks(rotation=45)
plt.show()
```
